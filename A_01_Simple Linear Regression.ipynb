```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\User\Desktop\Main\01_Data Science\Data\Advertising\Advertising.csv")
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('TV advertising')
plt.ylabel('Sales')
plt.scatter(df.TV,df.Sales)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['TV']],df.Sales)
```

```python
# Get Value of coefficient
print(reg.coef_)
print(reg.intercept_)
```

```python
# Validate results with function
c = reg.intercept_
m = reg.coef_
x = 225
y = m*x + c
print(y)
```

```python
reg.predict([[225]])
```

Trying predicting for a list of values

```python
spends = [223,220,105,12]
df_spends = pd.DataFrame(spends)

df_spends['sales'] = reg.predict(df_spends)
df_spends
```
