```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
```

```python
df = pd.read_csv("A_02_canada_per_capita_income.csv")
df.tail(5)
```

Here, we would like to predict for 2020, what is per capita income in canada (in USD)

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df['year'],df['per capita income (US$)'])
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
# Get Value of coefficient
print(reg.coef_)
print(reg.intercept_)
```

```python
# Validate results with function
c = reg.intercept_
m = reg.coef_
x = 2020
y = m*x + c
print(y)
```

```python
reg.predict([[2020]])
```

Lets see how good is the fit

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income')
plt.scatter(df.year,df['per capita income (US$)'])
plt.plot(df.year,reg.predict(df[['year']]),color = 'red')
```
